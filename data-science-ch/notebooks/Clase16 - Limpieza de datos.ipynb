```python
# Leer el dataset
url = 'https://raw.githubusercontent.com/JJTorresDS/ds-data-sources/main/income_per_person_gdppercapita_ppp_inflation_adjusted.csv'
df = pd.read_csv(url, sep=',')
df
```

## Verificacion de duplicados

```python
# Verificando si existen valores duplicados 
print("Registros iniciales: ", str(df.shape))
df = df.drop_duplicates()
print("Registros después de eliminar duplicados: ", str(df.shape))
```

## Comprensión de la estructura de la data (inicial)

```python
# Parte 1
# Recortamos la información filtrando por país = Argentina
# Recortamos columnas 217 a 224
df_pais = df[df['country'] == 'Argentina'].iloc[:,217:224]
lista = df_pais.values[0]
lista
```

```python
# Columnas recortadas
df_pais.columns
```

```python
# Obteniendop el promedio de GDP
lista_nueva = []
for string in lista:
  # Convirtiendo a formato numérico
  x = float(string.replace("k", ""))
  lista_nueva.append(x)
```

```python
statistics.mean(lista_nueva) * 1000
```

# 1. Calcular el GPD promedio de su país durante el periodo 2015-2021 y compartirlo con sus compañeros

```python
# Función para obtener gdp de país ingresado
def gdppc_promedio(pais, df_pais):
  df_pais = df[df['country'] == pais].iloc[:,217:224]
  lista = df_pais.values[0]
  lista_nueva = []
  for string in lista:
    # Convirtiendo a formato numérico
    x = float(string.replace("k", ""))
    lista_nueva.append(x)
  return(statistics.mean(lista_nueva) * 1000)
```

```python
gdppc_promedio('Argentina', df_pais)
```

```python
gdppc_promedio('Colombia', df_pais)
```

# 2. Calcular el GPD promedio de Sudamerica y Norteamerica en el mismo periodo de tiempo 

```python
# Parte 2 (Una forma de hacerlo) uso de filtros / Sudamérica
df_paises = df[df['country'].isin(['Argentina','Chile','Colombia','Bolivia','Peru','Brasil','Uruguay','Venezuela','Paraguay','Ecuador'])].iloc[:,217:224]
lista = df_paises.values
lista
```

```python
lista_nueva = []
for string in lista:
  for j in string:
    if 'k' in j:
      #print(j)
      x = float(j.replace("k", "")) * 1000
      lista_nueva.append(x)
      #print(x)
    else:
      x = float(j)
      lista_nueva.append(x)
      #print('---')
      #print(x)

statistics.mean(lista_nueva)
```

```python
# Paises presentes en el dataset
df.country.unique()
```

```python
# Norteamérica
df_paises = df[df['country'].isin(['United States','Canada','Mexico','Costa Rica','Nicaragua'])].iloc[:,217:224]
lista = df_paises.values
lista
```

```python
lista_nueva = []
for string in lista:
  for j in string:
    if 'k' in j:
      #print(j)
      x = float(j.replace("k", "")) * 1000
      lista_nueva.append(x)
      #print(x)
    else:
      x = float(j)
      lista_nueva.append(x)
      #print('---')
      #print(x)

statistics.mean(lista_nueva)
```
