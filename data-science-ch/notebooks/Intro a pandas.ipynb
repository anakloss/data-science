## Series

```python
import pandas as pd
import numpy as np
```

```python
numeros = range(50, 70, 2)
list(numeros)
```

```python
# Crear serie
num_serie = pd.Series(numeros)
num_serie
```

```python
# Acceder a atributos de la serie
num_serie.index
```

```python
num_serie.values
```

```python
# Crear serie con index pasado por parametro
num_en_texto = ['primero', 'segundo', 'tercero', 'cuarto', 'quinto', 'sexto',
                'séptimo', 'octavo', 'noveno', 'décimo']
num_serie_2 = pd.Series(numeros, index=num_en_texto)
num_serie_2
```

### Manipular datos en series

```python
num_serie_2['séptimo']
```

```python
num_serie_2.loc['séptimo']
```

## DataFrame

```python
# Crear dataframe a partir de series
modelos = ['A4 3.0 Quattro 4dr manual', 'A4 3.0 Quattro 4dr auto', 'A6 3.0 4dr',
           'A6 3.0 Quattro 4dr', 'A4 3.0 convertible 2dr']
```

```python
peso = [3583, 3627, 3561, 3880, 3814]
precios = ['$33,430', '$34,480', '$36,640', '$39,640', '$42,490']
```

```python
# creando series
autos_peso = pd.Series(peso, index=modelos)
autos_precio = pd.Series(precios, index=modelos)
```

```python
autos_peso
```

```python
autos_precio
```

```python
# creando dataframe
autos = pd.DataFrame({'Peso': autos_peso, 'Precio': autos_precio})
autos
```

### Manipular datos en Dataframe

```python
autos.index
```

```python
autos.columns
```

```python
autos.values
```

```python
autos['Peso']
```

```python
autos[['Peso']]
```

```python
autos.values[1]
```

```python
#por indice
autos.loc['A4 3.0 Quattro 4dr manual']
```

```python
# filtrar
autos.loc[autos.Peso >= 3600, 'Precio']
# autos.Peso = autos['Peso']
```

### Ejemplo: Construyendo un tablero de ajedrez con pandas y numpy

```python
# rango de 1 a 65
ajedrez_64 = np.arange(1, 65).reshape(8, 8)
ajedrez_64
```

```python
# creando dataframe
ajedrez_df = pd.DataFrame(ajedrez_64, columns=range(1, 9), index=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])
ajedrez_df
```
