# ANALISIS MULTIVARIADO

* Es una rama de la estadstica que abarca la observacin y el anlisis simultneos de ms de una variable respuesta.
* Los estudios multivariados son similares a los univariados y bivariados con la diferencia de que tienen ms de dos variables independientes y en algunos casos mltiples variables dependientes.
* Otra diferencia importante es que en un anlisis de mltiples variables no hablamos de correlacin simple ni de estadsticos descriptivos por s solos, sino que apelamos a otras herramientas estadsticas llamadas multivariantes, tal es el caso de por ejemplo: Anlisis de varianza (ANOVA), Estudio multifactorial o Regresiones Mltiples.

Los investigadores emplean **estudios multivariantes** cuando requieren examinar la relacin entre mltiples factores al mismo tiempo. Se diferencia claramente de los estudios univariados y bivariados en que **plantean ms de una variable dependiente y varias independientes**.

```python
#Importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
df = pd.read_csv(url)

df.columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
df.head()
```

```python
#Reemplazar el Outcome de 1 a 'Diab' y 0 a 'Non-Diab'
df.Outcome = df.Outcome.replace({0:'Non-Diab', 1:'Diab'})
df.head()
```

```python
#Renombrar la columna DiabetesPedigreeFunction a DPF
df.DiabetesPedigreeFunction = df.rename({'DiabetesPedigreeFunction': 'DPF'}, inplace=True, axis=1)
df.head()
```

```python
#Verificamos los tipos de datos
df.dtypes
```

```python
#Convertir a la variable: DPF en Integer y verificar su correcta conversion
df["DPF"] = pd.to_numeric(df["DPF"])
```

```python
#Validacion cambio de dato
df.dtypes
```

```python
#Shape
df.shape
```

```python
#Info
df.info()
```

```python
#Estadisticos bsicos
df.describe().T
```

```python
#Crear un Pairplot del dataset
plt.figure(dpi=120)
sns.pairplot(df)
plt.show()
```

```python
#Crear el Pairplot pero ahora con el atributo hue = 'Outcome'. Las leyendas debern decir: 'Non Diabetic','Diabetic'
plt.figure(dpi=120)
sns.pairplot(df, hue='Outcome', palette='plasma')
plt.legend(['Non Diabetic', 'Diabetic'])
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.heatmap(df.corr(), annot=True, cmap='Wistia')
plt.title('Mapa de calor', fontsize=15)
plt.show()
```

```python
#Correlaciones
plt.figure(dpi=120, figsize=(5,4))
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
sns.heatmap(df.corr(), mask=mask, fmt=".2f", annot=True, lw=1, cmap='plasma')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.jointplot(x='Glucose', y='Insulin', data=df, kind='scatter')
plt.show()
```

```python
sns.jointplot(x='Glucose', y='Insulin', data=df, kind='scatter', hue='Outcome')
plt.show()
```

```python
#JoinPlot
plt.figure(dpi=100, figsize=(5, 4))
print("Joint plot de Glucosa con otras variables ==> \n")
for i in  df.columns:
    if i != 'Glucose' and i != 'Outcome':
        print(f"Correlacion entre Glucose y {i} ==> ",df.corr().loc['Glucose'][i])
        sns.jointplot(x='Glucose',y=i,data=df,kind = 'scatter',color = 'purple')
        plt.show()
```

**Insight**: La glucosa muestra una asociacin lineal dbil positiva con las otras variables del conjunto de datos. Eso significa que al aumentar el nivel de glucosa en los pacientes, tambin aumentarn las otras variables.

```python
#Crear el join plot entre BloodPressure y el resto de las variables con su correspondiente interpretacin.

col = list(df.columns)
idx = col.index('BloodPressure')

plt.figure(dpi = 100, figsize = (5,4))
print("Joint plot de BloodPressure con otras variables ==> \n")
for i in  range(idx+1,len(col)-1):
    print(f"Correlacion entre BloodPressure y {col[i]} ==> ",df.corr().loc['BloodPressure'][col[i]])
    sns.jointplot(x='BloodPressure',y=col[i],data=df,kind = 'scatter',color = 'green')
    plt.show()
```

**Insight**: BloodPressure muestra una asociacin lineal dbil positiva con las otras variables del conjunto de datos dado. Eso significa que al aumentar el nivel de BP en los pacientes, tambin aumentarn el resto de las variables.

```python
#Crear el join plot entre SkinThickness y el resto de las variables con su correspondiente interpretacin.

col = list(df.columns)
idx = col.index('SkinThickness')

plt.figure(dpi = 100, figsize = (5,4))
print("Joint plot de SkinThickness con otras variables ==> \n")
for i in  range(idx+1,len(col)-1):
    print(f"Correlacion entre SkinThickness y {col[i]} ==> ",df.corr().loc['SkinThickness'][col[i]])
    sns.jointplot(x='SkinThickness',y=col[i],data=df,kind = 'reg',color = 'blue')
    plt.show()
```

**Insight**: SkinThickness muestra una asociacin lineal dbil positiva con el resto de las variables, excepto con la edad. Eso significa que al aumentar el SkinThickness en los pacientes, las otras variables tambin aumentarn. SkinThickness con la edad muestran una correlacin negativa dbil es decir, al aumentar SkinThickness, la edad disminuir.
