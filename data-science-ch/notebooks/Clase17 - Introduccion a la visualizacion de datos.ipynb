# VISUALIZACION DE DATOS



Se refiere a la presentacin de datos en un formato ilustrado o grfico. \
Permite **observar** el anlisis de datos, de manera visual, para captar **conceptos difciles o identificar patrones**.

## Para qu sirve?
Gracias a la visualizacin de datos, las personas encargadas de tomar decisiones por ejemplo dentro de una empresa, ven significativamente facilitada su labor, al contar con medios visuales que simplifican la comprensin del conjunto de datos.


```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Omitir warnings
import warnings
warnings.filterwarnings('ignore')
```

# Scatterplot

```python
# Lectura de csv
df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv')
df.head()
```

```python
df.rename(columns={'Unnamed: 0': 'CarName'}, inplace=True)
df.head()
```

```python
# tamao de figura
fig = plt.figure(figsize=(12, 8))
# Grafico: linea de puntos y regresin
ax = sns.regplot(x="wt", y="mpg", data=df)
ax.set(xlabel='MPG', ylabel='WT')
plt.title('Peso vs mpg')
plt.show()
```

# Barplot

```python
df_grpd = df.cyl.value_counts()
# Usando groupby
# df_grpd = df.groupby('cyl').count().reset_index()
df_grpd
```

```python
# tamao de figura
fig = plt.figure(figsize=(12, 8))
# Grafico: linea de puntos y regresin
ax = sns.barplot(x=df_grpd.index, y=df_grpd.values)
ax.set(xlabel='N de carros por cada cilindrada', ylabel='Cilindros')
plt.title('Cilindradas')
plt.show()
```

# Boxplot

```python
fig = plt.figure(figsize=(12, 8))
sns.boxplot(x="vs", y='wt', data=df)
plt.title('Vs vs WT')
plt.show()
```

# Histograma

```python
df = pd.read_csv('https://raw.githubusercontent.com/marsja/jupyter/master/flanks.csv', index_col=0)
df.head()
```

```python
# Grafico
fig = plt.figure(figsize=(12, 8))
for condition in df.TrialType.unique():
    cond_data = df[(df.TrialType == condition)]
    ax = sns.distplot(cond_data.RT, kde=False, label=condition)
ax.set(xlabel='Tiempo de respuesta', ylabel='Frecuencia')
ax.legend()
ax.set_title('Histogramas Tiempo de respuesta')
```

# Series de tiempo

```python
df = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-02-26/full_trains.csv")
df.head()
```

```python
plt.figure(figsize=(12, 8))
# desagregar por estacin que contenga 'PARIS'
sns.lineplot(x="month", y="total_num_trips", hue="departure_station", ci=None,  # ci es el sombreado de la linea
             data=df[df.departure_station.str.contains('PARIS')])
plt.xlabel('Mes')
plt.ylabel('Numero de viajes')
plt.title('Viajes por mes')
plt.show()
```

```python
import pandas as pd
import ptitprince as pt
```

```python
df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/iris.csv')
df.head()
```

```python
plt.figure(figsize=(12, 8))
ax = pt.RainCloud(x='Species', y='Sepal.Length', data=df, 
                  width_viol=.8, width_box=.4, orient='h', move=.0)
plt.show()
```

De este grfico podemos ver que los grficos se separan por categoras. 

```python
plt.figure(figsize=(12, 8))
ax = pt.RainCloud(x='Species', y='Petal.Length', data=df, 
                  width_viol=.8, width_box=.4, orient='h', move=.0)
plt.show()
```
