```python
# Concatenacion
# axis=0 concatena por filas y axis=1 concatena por columnas
array_1 = np.random.randint(10, size=5)
array_2 = np.random.randint(10, size=5)

arrays_conc = np.concatenate((array_1, array_2))
arrays_conc
```

```python
array_extra = np.array([[10],[20]])
array_extra
```

```python
# vstack y hstack para apilar
arrays_apilados_v = np.vstack([array_extra, array_extra])
arrays_apilados_v
```

```python
# split, vsplit y hsplit para desarmar o partir
array_partido = np.split(arrays_conc, [4, 6])
array_partido
```

### Calculo con funciones utiles

```python
calificaciones = np.array([10, 10, 8, 9, 10])

# promedio
calificaciones.mean()
```

```python
# suma
calificaciones.sum()
```

```python
# maximo
calificaciones.max()
```

```python
# media
np.median(calificaciones)
```

```python
# desviacion estandar
np.std(calificaciones)
```
