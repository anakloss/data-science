## Parámetros de la función [read_csv](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)
```
pd.read_csv(filepath="python-ml-course/datasets/titanic/titanic3.csv",
    sep=",",
    dtype={'ingresos': np.float64, 'edad': np.int32},
    header=0,
    names={"ingresos", "edad"},
    skiprows=12,
    index_col=None,
    skip_blank_lines=False,
    na_filter=False)
```
* dtype = convertir columna a tipo de dato necesario(np = libreria numpy).
* header = expecificar fila de cabecera. Puede usarse un int o lista.
* names = formatea los nombres de cabecera. Se deben pasar todos los nombres de las columnas.
* skiprows = saltar filas (Ej: se carga desde la fila 13).
* index_col = formatea columna del dataset con la informacion pasada.
* skip_blank_lines = saltear las lineas con valores vacios o en blanco.
* na_filter = elimina filas con valores vacios o NaN.

# Carga de datos a través de la función [open](https://www.w3schools.com/python/ref_func_open.asp)

#### Funciones
* next = siguiente linea
* strip = eliminar espacios en blanco al inicio o final de la linea
* split = dividir por un delimitador

````python
data3 = open(mainpath + "/customer-churn-model/Customer Churn Model.txt", "r")
````

````python
cols = data3.readline().strip().split(",")
n_cols = len(cols)
````

````python
count = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
````

````python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    count += 1

print("El data set tiene %d filas y %d columnas" % (count, n_cols))
````

````python
# Convertir el diccionario creado con open a DataFrame
df3 = pd.DataFrame(main_dict)
df3.head()
````

## Lectura y escritura de ficheros

````python
infile = mainpath + "/customer-churn-model/Customer Churn Model.txt"
outfile = mainpath + "/customer-churn-model/Tab Customer Churn Model.txt"
````

````python
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
````

````python
# Muestra de fichero escrito
df4 = pd.read_csv(outfile, sep="\t")
df4
# df4.head()
````

### Ejemplo de descarga de datos desde URL y decodificación de archivo Byte
Utilizando libreria urllib3 utilizada para acceder a información desde una URL, y procesando los datos para guardarlos en CSV o TXT.

````python
import urllib3
import csv

http = urllib3.PoolManager()
r = http.request("GET", medals_url)  # Peticion
r.status
response = r.data.decode('ascii')  # Datos

# Hecho anteriormente en la funcion open
cr = csv.reader(response)
cr
````

````python
outfile = "../python-ml-course-output/winter olympics medals.txt"
with open(outfile, "w") as outfile1:
    for line in response.split("\n"):
        fields = line.strip().split(",")
        outfile1.write("\t".join(fields))
        outfile1.write("\n")

wom = pd.read_csv(outfile, sep="\t")
wom.head()
````
