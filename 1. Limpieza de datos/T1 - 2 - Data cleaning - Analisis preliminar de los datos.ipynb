# Variables Dummy

Las variables dummy o indicadoras sirven para identificar categorias o clase a las que pertenecen las observaciones.

```python
data['sex'].tail()
```

```python
dummy_sex = pd.get_dummies(data['sex'], prefix="sex")
```

```python
dummy_sex.tail()
```

```python
# Adjuntar dummy al dataset
column_names = data.columns.values.tolist()
column_names
```

```python
# Eliminar columna "sex"
data = data.drop(['sex'], axis=1).head()
```

```python
# Agrega dummy "sex"
data = pd.concat([data, dummy_sex], axis=1)
```

```python
data.head()
```

```python
# Función para concatenar dummys en archivo original
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis=1)
    df = pd.concat([df, dummy], axis=1)
    return df
```
